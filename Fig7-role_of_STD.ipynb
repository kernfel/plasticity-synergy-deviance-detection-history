```python
import matplotlib.pyplot as plt
import numpy as np
from brian2.only import *
import seaborn as sns
import deepdish as dd
from scipy.stats import bootstrap

import plot_utils as pu
from digest import conds, get_digest_output, get_voltages
import isi_indexing as ii
from readout import load_results
from digest import get_voltage_histograms

import conf.isi5_mouse as cfg
from demonstrator import demonstrator
```

```python
labelling = {
    'pspike': 'Spike probability',
    'Depression': '$V_{STD}$ (mV)',
    'Threshold': '$V_{TA}$ (mV)',
    'Synapses': '$V_{syn}$ (mV)',
    'Reset': '$V_{reset}$ (mV)',
    'v': 'Membrane potential v (mV)',
    'u': 'Undepressed potential u (mV)',
    'th_adapt': 'Threshold (mV)',
    'vsyn': '$v_{syn}$',
    'Vall': '$\\Sigma V$ (mV)'
}
depressive_measures = ('Depression', 'Threshold')
voltage_measures = ('Reset', 'Synapses', 'Depression')
raw_measures = 'th_adapt', 'u', 'v', 'vsyn'
pu.colors.update({'Activity': 'cyan', 'Vall': 'k',
                  'th_adapt': pu.colors['Threshold'], 'u': pu.colors['Depression'], 'v': 'b', 'vsyn':'magenta'})
conds = ('std', 'msc', 'dev')
nonstandard = ('msc', 'dev')
```

```python
isi, STD, TA, templ = 500, 1, 0, 'R'
net, ipair, istim = [demonstrator[k] for k in ('net', 'ipair', 'istim')]
pair, stim = cfg.pairings[ipair], cfg.pairings[ipair][istim]
res = load_results(cfg.fname.format(net=net, isi=isi, STD=STD, TA=TA, templ=templ), {'th_adapt': 0}, compress=True, tmax=100*ms)

res['voltage_histograms'] = get_voltage_histograms(cfg.params, res)
```

```python
quarters = [.25, .5, .75]
```

```python
histograms = {cond: {
    'pspike': res['spikes'][ipair][stim][cond]['spike_hist'],
    **{measure: res['voltage_histograms'][measure][ipair][stim][cond] for measure in voltage_measures}
} for cond in conds}
```

```python
nspikes = {cond: res['spikes'][ipair][stim][cond]['nspikes'] for cond in conds}
```

```python
raw_voltages = {cond: {
    **get_voltages(cfg.params, {k:v*volt for k,v in res['dynamics'][ipair][stim][cond].items()}),
    **{k:v*volt for k,v in res['dynamics'][ipair][stim][cond].items()},
} for cond in conds}
for v in raw_voltages.values():
    v['Vall'] = cfg.params['v_rest'] - cfg.params['v_threshold'] + v['Synapses'] + v['Reset'] - v['Depression'] - v['Threshold']
    assert np.allclose(v['Vall']/mV, (v['v'] - (cfg.params['v_threshold']+v['th_adapt']))/mV)
```

```python
index_N, index_t = ii.get_onset_ordering(cfg, np.stack([h['pspike'] for h in histograms.values()]))
tmax = np.max(index_t)+1
time = np.arange(tmax) * cfg.params['dt']
```

# A: Deviance detection happens with TA-only

```python
fig, ax = plt.subplots(figsize=(4.5,3))
for cond, hists in histograms.items():
    pspike = hists['pspike'].sum(0)[:tmax]
    ax.plot(time/ms, pspike, c=pu.colors[cond], label=cond)
plt.xlabel('Time after stimulus (ms)')
plt.ylabel('Mean spikes per pulse')
sns.despine()

# TODO: Inset with DD/SSA index across nets
# NOTE: Ensure same template in Fig2
```

```python
pdiff = histograms['dev']['pspike'] - histograms['msc']['pspike']
psign_contrast = {}
masks = (pdiff > 0.05, pdiff < -0.05)
for measure in voltage_measures:
    contrast = histograms['dev'][measure] - histograms['msc'][measure]
    contrast = np.stack([contrast * np.where(mask, 1, np.nan) for mask in masks])
    psign_contrast[measure] = contrast
```

```python
fig, axs, cb = pu.plot_pulse_hist([np.where(sum(masks), pdiff, 0)], index_N, index_t, cfg.params['dt'], figsize=(5,8), cmap=pu.ccc_divergent)
cb.set_label('$\\Delta p_{spike}$')

fig, axs, cb = pu.plot_pulse_hist(
    [(histograms['dev']['Depression'] - histograms['msc']['Depression'])/mV],
    index_N, index_t, cfg.params['dt'], figsize=(5,8), cmap=pu.ccc_divergent)
cb.set_label('$\\Delta V_{STD}$ (mV)')

for measure, contrast in psign_contrast.items():
    fig, axs, cb = pu.plot_pulse_hist(contrast/mV, index_N, index_t, cfg.params['dt'], figsize=(10,8))
    cb.set_label(labelling[measure])

pu.plot_pulse_hist(
    np.abs(psign_contrast['Depression']) / sum(np.abs(contrast) for contrast in psign_contrast.values()),
    index_N, index_t, cfg.params['dt'], figsize=(10,8))
```

```python
fig, axs = plt.subplots(1,2, figsize=(15,5), sharex=True)
time = np.arange(tmax) * cfg.params['dt'] / msecond
axs[0].set_title('Bins with more activity in deviant')
axs[1].set_title('Bins with more activity in control')
plt.suptitle('Unweighted')
for measure, contrast in psign_contrast.items():
    for i, ax in enumerate(axs):
        q = np.nanquantile((contrast[i])[:, :tmax]/mV, quarters, 0)
        ax.fill_between(time, q[0], q[2], color=pu.colors[measure], alpha=.2)
        ax.plot(time, q[1], color=pu.colors[measure], label=labelling[measure])
for ax in axs:
    ax.set_xlabel('Time (ms)')
    ax.set_ylabel('$\\Delta V$ (mV)')
axs[0].legend()
sns.despine()
```

```python
fig, axs = plt.subplots(1,2, figsize=(15,5), sharex=True)
time = np.arange(tmax) * cfg.params['dt'] / msecond
axs[0].set_title('Bins with more activity in deviant')
axs[1].set_title('Bins with more activity in control')
plt.suptitle('Weighted by $|\\Delta p_{spike}|$')
for measure, contrast in psign_contrast.items():
    for i, ax in enumerate(axs):
        q = np.nanquantile((contrast[i]*np.abs(pdiff))[:, :tmax]/mV, quarters, 0)
        ax.fill_between(time, q[0], q[2], color=pu.colors[measure], alpha=.2)
        ax.plot(time, q[1], color=pu.colors[measure], label=labelling[measure])
for ax in axs:
    ax.set_xlabel('Time (ms)')
    ax.set_ylabel('Weighted $\\Delta V$ (a.u.)')
axs[0].legend()
sns.despine()
```

```python
fig, axs = plt.subplots(1,2, figsize=(15,5), sharex=True)
axs[0].set_title('Bins with more activity in deviant')
axs[1].set_title('Bins with less activity in deviant')
for i, ax in enumerate(axs):
    ratio = np.abs(psign_contrast['Depression'][i]) / sum([np.abs(contrast[i]) for contrast in psign_contrast.values()])
    q = np.nanquantile((ratio)[:, :tmax], quarters, 0)
    ax.fill_between(time, q[0], q[2], alpha=.2, color=pu.colors['Depression'])
    ax.plot(time, q[1], color=pu.colors['Depression'])
for ax in axs:
    ax.set_xlabel('Time (ms)')
    ax.set_ylabel('Fraction of $\\Delta V$ contributed by $V_{STD}$')
sns.despine()
```

```python
fig, axs = plt.subplots(1,2, figsize=(15,5), sharex=True)
time = np.arange(tmax) * cfg.params['dt'] / msecond
axs[0].set_title('Bins with more activity in deviant')
axs[1].set_title('Bins with more activity in control')
plt.suptitle('Weighted by $|\\Delta p_{spike}|$')
measure = 'Depression'
for i, ax in enumerate(axs):
    q = np.nanquantile((psign_contrast[measure][i]*np.abs(pdiff))[:, :tmax]/mV, quarters, 0)
    ax.fill_between(time, q[0], q[2], color=pu.colors[measure], alpha=.2)
    ax.plot(time, q[1], color=pu.colors[measure], label=labelling[measure])
for ax in axs:
    ax.set_xlabel('Time (ms)')
    ax.set_ylabel('$\\Delta V$ (mV)')
axs[0].legend()
sns.despine()
```

```python
fig, axs = plt.subplots(1,2, figsize=(15,5), sharex=True)
time = np.arange(tmax) * cfg.params['dt'] / msecond
plt.suptitle('Bins with more activity in deviant')
axs[0].set_title('Voltages weighted by $\\Delta p_{spike}$')
axs[1].set_title('Relative contribution')
for measure in ('Threshold', 'Depression', 'Synapses'):
    if measure not in psign_contrast:
        continue
    q = np.nanquantile((psign_contrast[measure][0]*np.abs(pdiff))[:, :tmax]/mV, quarters, 0)
    axs[0].fill_between(time, q[0], q[2], color=pu.colors[measure], alpha=.2)
    axs[0].plot(time, q[1], color=pu.colors[measure], label=labelling[measure])
axs[0].set_ylabel('Weighted $\\Delta V$ (a.u.)')

for measure in ('Threshold', 'Depression'):
    if measure not in psign_contrast:
        continue
    ratio = np.abs(psign_contrast[measure][0]) / sum([np.abs(contrast[0]) for contrast in psign_contrast.values()])
    q = np.nanquantile((ratio)[:, :tmax], quarters, 0)
    axs[1].fill_between(time, q[0], q[2], alpha=.2, color=pu.colors[measure])
    axs[1].plot(time, q[1], color=pu.colors[measure])
axs[1].set_ylabel('Fraction of total $\\Delta V$')

for ax in axs:
    ax.set_xlabel('Time (ms)')
axs[0].legend()
sns.despine()
```

```python
res_TA = load_results(cfg.fname.format(net=net, isi=isi, STD=0, TA=1, templ=templ), {'u': 'v'}, compress=True, tmax=100*ms)
res_TA['voltage_histograms'] = get_voltage_histograms(cfg.params, res_TA)
voltage_measures_TA = ('Threshold', 'Synapses', 'Reset')
```

```python
histograms_TA = {cond: {
    'pspike': res_TA['spikes'][ipair][stim][cond]['spike_hist'],
    **{measure: res_TA['voltage_histograms'][measure][ipair][stim][cond] for measure in voltage_measures_TA}
} for cond in conds}
```

```python
pdiff_TA = histograms_TA['dev']['pspike'] - histograms_TA['msc']['pspike']
psign_contrast_TA = {}
masks = (pdiff_TA > 0.05, pdiff_TA < -0.05)
for measure in voltage_measures_TA:
    contrast = histograms_TA['dev'][measure] - histograms_TA['msc'][measure]
    contrast = np.stack([contrast * np.where(mask, 1, np.nan) for mask in masks])
    psign_contrast_TA[measure] = contrast
```

```python
fig, axs = plt.subplots(1,2, figsize=(15,5), sharex=True)
time = np.arange(tmax) * cfg.params['dt'] / msecond
plt.suptitle('Bins with more activity in deviant: TA-only vs STD-only')
axs[0].set_title('Voltages weighted by $\\Delta p_{spike}$')
axs[1].set_title('Relative contribution')
for measure, contr, diff in (['Threshold', psign_contrast_TA, pdiff_TA], ['Depression', psign_contrast, pdiff]):
    weighted_dv = (contr[measure][0]*np.abs(diff))[:, :tmax]
    q = np.nanquantile(weighted_dv/mV, quarters, 0)
    axs[0].fill_between(time, q[0], q[2], color=pu.colors[measure], alpha=.2)
    axs[0].plot(time, q[1], color=pu.colors[measure], label=f'Only {labelling[measure].split()[0]} (total: {np.nansum(weighted_dv)*volt/mV:.2f})')

    ratio = np.abs(contr[measure][0]) / sum([np.abs(contrast[0]) for contrast in contr.values()])
    q = np.nanquantile((ratio)[:, :tmax], quarters, 0)
    axs[1].fill_between(time, q[0], q[2], alpha=.2, color=pu.colors[measure])
    axs[1].plot(time, q[1], color=pu.colors[measure], label=f'Only {labelling[measure].split()[0]} (mean: {np.nanmean(ratio):.2f})')
axs[0].set_ylabel('Weighted $\\Delta V$ (a.u.)')
axs[1].set_ylabel('Fraction of total $\\Delta V$')

for ax in axs:
    ax.set_xlabel('Time (ms)')
    ax.legend()
sns.despine()
```
